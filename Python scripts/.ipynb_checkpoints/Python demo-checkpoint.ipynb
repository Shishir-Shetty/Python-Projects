```python
#Creating Dummy dictionary. Creating list in dictionary
Details = { 
    'Name' : ['shishirshetty','Shir','Shir'],
    'Email' : ['shishirshetty@gmail.com','Shir@gmail.com','Shetty@gmail.com'],
    'Phone':[100,101,102]
}
#Assigning Dictionary to new Df
import pandas as pd
newdf = pd.DataFrame(Details)
```

```python
#'iloc' is used to access the particular by giving the Index of the row inside '[]'. iloc stands for index location
newdf.iloc[0]
```

```python
#printing 2 rows together
newdf.iloc[[0,1]]
```

```python
#we can print the specified rows and columns using their indexes. [[0,1],1], where[0,1] is the row index and 1 is the column index
newdf.iloc[[0,1],1]
```

```python
#In 'iloc' we use index to print rows and columns, but in 'loc' we used index for rows and labels for columns.
#we can print the specified rows and columns using their indexes. [[0,1],'Email'], where[0,1] is the row index and 'Email' is the column label
newdf.loc[[0,1],'Email']
```

```python
#printing 2 rows and 2 columns together.[0,1] is the row index.['Name','Email'] is the column label
newdf.loc[[0,1],['Name','Email']]
```

```python
#printing 2 rows together by using list slicer i.e,0:1. when you use slicer dont make us of '[]'
newdf.loc[0:1,['Name']]
```

```python
#printing 2 columns together by using list slicer i.e, 'Name':'Phone'. when you use slicer dont make us of '[]'
newdf.loc[0:1,'Name':'Phone']
```

```python
#Filtering. First assign the filter condition to one of the variale('filt') and then print it
filt = (newdf['Name']=='Shir')
newdf[filt]
#Also can pass the filter variable to 'loc' to get the particular column.Below code give result with only 'Phone' Column.
newdf.loc[filt,'Phone']
```

```python
#AND operator '&'
filt = (newdf['Name']=='Shir') & (newdf['Phone']==101)
newdf.loc[filt,'Email']

#OR operator  '|'
filt = (newdf['Name']=='Shir') | (newdf['Phone']==101)
newdf.loc[filt,'Email']

#NOT operator '`'
filt = (newdf['Name']=='Shir')
newdf.loc[~filt,'Phone']
```

```python
#'isin' method is used like 'is present in the'. Create list of countries 'countries' and then apply filter for the dataframe using 'isin'
countries = ['United States','India','United Kingdom']
filt = (df['Country'].isin(countries))
df.loc[filt,'Country']
```

```python
#'str.contains' is one of the string method.'Python' is the string we are matching and 'na' Should be given as false since there might be no values sometime.
filt = (df['LanguageWorkedWith'].str.contains('Python',na= False))
df.loc[filt,['Country','LanguageWorkedWith']]
```

```python
#to convert all the columns to Upper case
newdf.columns = (x.upper() for x in newdf.columns)
#to convert all the columns to lower case
newdf.columns = (x.lower() for x in newdf.columns)
```

```python
#'replace' is used to rename columns,remove space etc.
newdf.columns = newdf.columns.str.replace('phoNe','Phone')
```

```python
#Rename can also done by below method
newdf.rename(columns = {'Name':'NaMe'},inplace = True)
```

```python
#to change Values of particular column of particular row can be done by below method
newdf.loc[[2],['Name','Phone']]=['Shishir',103]
newdf
```

```python
#convert values in the 'Email' column to lower case. To convert into upper case replace 'lower' with 'upper'
newdf['Email'] = newdf['Email'].str.lower()
newdf
```

```python
newdf.loc[2,['Name']]='Shishir'
newdf
```

```python
#to find the length of all the rows of 'Email' column.
newdf['Email'].apply(len)
```

```python
#Another method to convert into Upper or lower
#Create a function 'Upper_Email' which takes one argument and converts into upper
def Upper_Email(email):
    return email.upper()
#pass the Function like below.It will convert all the rows in'Email'column to upper.
newdf['Email'] = newdf['Email'].apply(Upper_Email)
```

```python
#Another method to convert into Upper or lower using Lambda method
newdf['Email']=newdf['Email'].apply(lambda x:x.lower())
```
