## Imports

```python
import pandas as pd
import glob
```

## Check for any errors

```python
ls1 = glob.glob("t20s_csv2\*.csv")
ls2 = []

for i in ls1:
    try:
        ls2.append(pd.read_csv(i))
    except:
        print('error', i)
```

## Combine Match by Match CSV files

```python
#importing multiple files in a directory
l = [pd.read_csv(filename) if "_info" not in filename else print(filename) for filename in glob.glob("t20s_csv2\*.csv")]
df = pd.concat(l, axis = 0)
```

## Sneak peak

```python
df.tail()
```

## Sort by dates

```python
df['start_date'] = pd.to_datetime(df['start_date'])
```

```python
df = df.sort_values('start_date', ascending = True)
```
