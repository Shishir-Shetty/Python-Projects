```python
batsman = ['CA Pujara','V Kohli','RA Jadeja','AM Rahane','RG Sharma','KL Rahul','RR Pant','MA Agarwal']
```

```python
batsmandf = df[df['striker'].isin(batsman)]
```

```python
batsmandf['start_date'] =  pd.to_datetime (batsmandf['start_date'])
```

```python
batsmandf['Year'] = batsmandf['start_date'].dt.year
```

```python
def balls_per_dismissal(balls, dismissals):
    if dismissals > 0:
        return balls/dismissals
    else:
        return balls/1 
    
def balls_per_boundary(balls, boundaries):
    if boundaries > 0:
        return balls/boundaries
    else:
        return balls/1 
    
def runs_per_dismissal(runs,dismissals):
    if dismissals > 0:
        return runs/dismissals
    else:
        return runs/1
```

```python
#combining all into one Function

def playerStatistics(df):
    
    
#     df['isDot'] = df['runs_off_bat'].apply(lambda x:1 if x==0 else 0)
#     df['isOne'] = df['runs_off_bat'].apply(lambda x:1 if x==1 else 0)
#     df['isTwo'] = df['runs_off_bat'].apply(lambda x:1 if x==2 else 0)
#     df['isThree'] = df['runs_off_bat'].apply(lambda x:1 if x==3 else 0)
#     df['isFour'] = df['runs_off_bat'].apply(lambda x:1 if x==4 else 0)
#     df['isSix'] = df['runs_off_bat'].apply(lambda x:1 if x==6 else 0)
    
    runs = pd.DataFrame(df.groupby(['striker'])['runs_off_bat'].sum()).rename(columns ={"runs_off_bat":"runs"}).reset_index()
    balls = pd.DataFrame(df.groupby(['striker'])['match_id'].count()).rename(columns ={"match_id":"balls"}).reset_index()
    
    #we get innings by Matchid and striker
    innings = pd.DataFrame(batsmandf.groupby(['match_id','striker'])['innings'].apply(lambda x: len(list(set(x))))).reset_index()
    #we 
    innings = pd.DataFrame(innings.groupby(['striker'])['innings'].sum()).reset_index()
    
    dismissed = pd.DataFrame(df.groupby(['striker'])['player_dismissed'].count()).rename(columns ={"player_dismissed":"dismissals"}).reset_index()
    
#     dots = pd.DataFrame(df.groupby(['striker'])['isTwo'].sum()).rename(columns ={"isTwo":"dots"}).reset_index()
#     ones = pd.DataFrame(df.groupby(['striker'])['isOne'].sum()).rename(columns ={"isOne":"ones"}).reset_index()
#     twos = pd.DataFrame(df.groupby(['striker'])['isTwo'].sum()).rename(columns ={"isTwo":"twos"}).reset_index()
#     threes = pd.DataFrame(df.groupby(['striker'])['isThree'].sum()).rename(columns ={"isThree":"threes"}).reset_index()
#     fours = pd.DataFrame(df.groupby(['striker'])['isFour'].sum()).rename(columns ={"isFour":"fours"}).reset_index()
#     sixes = pd.DataFrame(df.groupby(['striker'])['isSix'].sum()).rename(columns ={"isSix":"sixes"}).reset_index()
    
#     df = pd.merge(innings,runs, on ='striker').merge(balls, on ='striker').merge(dismissed,on ='striker').merge(dots,on ='striker').merge(ones,on ='striker').merge(twos,on ='striker').merge(threes,on ='striker').merge(fours,on ='striker').merge(sixes,on ='striker')
    df = pd.merge(innings,runs, on ='striker').merge(balls, on ='striker').merge(dismissed,on ='striker')
    df['RPI'] = df.apply(lambda x:(x['runs']/x['innings']), axis=1)
    
    #Strike rate
#     df['SR'] = df.apply(lambda x:(100 * x['runs']/x['balls']), axis=1)
    #balls per dismissals
    df['BPD'] = df.apply(lambda x:balls_per_dismissal(x['balls'],x['dismissals']), axis=1)
    #runs per dismissals
    df['RPD'] = df.apply(lambda x:runs_per_dismissal(x['runs'],x['dismissals']), axis =1)
    #balls Per boundary
#     df['BPB'] = df.apply(lambda x: balls_per_boundary(x['balls'], (x['fours'] + x['sixes'])), axis = 1)
    return df
```

```python
df1 = playerStatistics(batsmandf)
```

```python
df1
```

```python
#Grouped the total runs together using match-id and striker
df2 = pd.DataFrame(batsmandf.groupby(['match_id','striker']).apply(lambda x: x['team_runs'].unique())).reset_index()
#renaming the empty column returned
df2.rename(columns = {	0 : 'Team_runs'}, inplace=True)
#converting 'team_runs' into integer
df2['Team_runs']= df2['Team_runs'].astype(int)
df2

```

```python
# Group using striker
df3 = pd.DataFrame(df2.groupby(['striker'])['Team_runs'].sum()).reset_index()
df3
```

```python
final_df = pd.merge(df1,df3, on ='striker')
final_df
```

```python
final_df['Contribution_To_Team'] = final_df['Team_runs']/final_df['runs']
final_df = final_df.sort_values(by=['Contribution_To_Team'], ascending=False)
final_df
```
