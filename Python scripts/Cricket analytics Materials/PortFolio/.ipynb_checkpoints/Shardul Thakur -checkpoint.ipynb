```python
%matplotlib inline
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)  
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', -1)
```

```python
deliveries = pd.read_csv("C:\\Users\\SHETTS60\\Documents\\Python scripts\\Cricket analytics Materials\\CricSheet\\Data Files\\TESTS_ball_by_ball.csv")
```

```python
df = deliveries.copy()
```

```python
df = df[df['bowling_team'] == 'India']
```

```python
df.head(2)
```

```python
#Converting 'start_date' into datetime to fetch the year from it
df["start_date"] = pd.to_datetime(df["start_date"])
```

```python
df['year'] = df["start_date"].dt.year
```

```python
df.head(2)
```

```python
df["year"].unique()
```

```python
#Assuming Takur performed for india from 2019
year_filter = [2020,2021,2022]
```

```python
df = df[(df["year"].isin(year_filter))]
```

```python
df.head(2)
```

```python
# if value is 'nan' then type will be float. Other than nan type will be str, hence we will consider only value which has str type
def is_bowlers_wicket(player_dismissed, dismissal_kind):
    if type(player_dismissed ) == str:
        if dismissal_kind not in [ 'run out', 'retired hurt', 'obstructing the field']:
            return 1
        else:
            return 0
    else:
        return 0
```

```python
df["isBowlerWk"] = df.apply(lambda x: is_bowlers_wicket(x["player_dismissed"], x["wicket_type"]),axis =1)
```

```python
# calculating the total runs of a batsman
df1 = pd.DataFrame( df.groupby(["match_id","innings","striker","start_date"])["runs_off_bat"].sum().reset_index())
```

```python
# calculating the total balls of a batsman
df2 = pd.DataFrame(df.groupby(["match_id","innings","striker","start_date"])["ball"].count().reset_index())
```

```python
#Adding the bowler who dismissed."ball" have to be aggregated to count since groupby needs atleast one aggregation
df3= pd.DataFrame(df[df['isBowlerWk'] ==1].groupby(["match_id","innings","striker","start_date","bowler"])["ball"].count().reset_index())
```

```python
#Merging df1,df2,df3 into df4
df4 = pd.merge(df1,df2, on =["match_id","innings","striker","start_date"]).merge(df3, on =["match_id","innings","striker","start_date"])
```

```python
df4.head()
```

```python
#drop the ball_y and renaming ball_x to balls
df4 = df4.rename(columns ={"ball_x":"balls"}).drop('ball_y',axis=1)
```

```python
df4
```

```python
#Filtering >30 runs or >50 balls to define who has settled in crease 
df4 = df4[(df4["runs_off_bat"] > 30) | (df4["balls"] > 50) ]
```

```python
# No of wickets of settled batsman
df5 =pd.DataFrame( df4['bowler'].value_counts().reset_index().rename(columns ={"index":"bowler","bowler":"wickets"}))
df5
```

```python
#Total Number of matches of bowler
df6 =  pd.DataFrame(df.groupby(["bowler"])["match_id"].apply(lambda x: len(np.unique(x)))).reset_index().rename(columns ={"match_id":"matches"})
df6
```

```python
finaldf = pd.merge(df5,df6, on =["bowler"])
finaldf
```

```python
#Wickets Per matches of a setlled batsman
finaldf['WPM'] =finaldf['wickets']/finaldf['matches']
finaldf.sort_values("WPM" , ascending = False).reset_index(drop = True)
```

## Shardul has the highest Wicket per matches of a settled batsman in recent time.
## Hence Clearly shows Shardul takur provides major breakthrough
